```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib.pyplot as plt
from labellines import labelLines
from matplotlib import rcParams
import pandas as pd

# Define colors using a list
def plot_multiple_curves(x_list, y_list, color_indices, plot_title):
    colors = [
        '#00549f',  # 0: dark_blue
        '#407fb7',  # 1: medium_blue
        '#8ebae5',  # 2: light_blue
        '#c7ddf2',  # 3: sky_blue
        '#e8f1fa',  # 4: very_light_blue
        '#57ab27',  # 5: new_color_1 (green)
        '#8dc060',  # 6: new_color_2 (light green)
        '#b8d698',  # 7: new_color_3 (pale green)
        '#ddebce',  # 8: new_color_4 (very pale green)
        '#f2f7ec'   # 9: new_color_5 (almost white)
    ]

    # Set global font to Times New Roman
    rcParams['font.family'] = 'Times New Roman'
    rcParams['mathtext.fontset'] = 'stix'  # Use Times font for math
    rcParams['mathtext.rm'] = 'Times New Roman'  # Set Times for regular math text

    # Set figure size (half A4 width, 8 cm height)
    fig = plt.figure(figsize=(10.5 / 2.54, 10.5 / 2.54))  # Convert cm to inches
    # Remove the top and right spines (frames)
    ax = plt.gca()  # Get current axis
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Plot each curve in the list
    for i in range(len(x_list)):
        plt.plot(x_list[i], y_list[i], label=f"Curve {i+1}", color=colors[color_indices[i]])
    
    # Adding labels and legend
    plt.title(plot_title)
    plt.xlabel('X-axis label')
    plt.ylabel('Y-axis label')
    plt.legend(loc='best')
    plt.show()

```

```python
# Example data sets
index = 3500
x1 = df_wb1['V_Mitte [mm]'][:index]
y1 = df_wb1['KMD1 [kN]'][:index]

x3 = df_wb3['V_Mitte [mm]'][:index]  # Second set of data (example)
y3 = df_wb3['KMD1 [kN]'][:index]  # Second set of data (example)

x4 = df_wb4['V_Mitte [mm]'][:index]  # Second set of data (example)
y4 = df_wb4['KMD1 [kN]'][:index]  # Second set of data (example)

# Define the x and y datasets and the corresponding color indices
x_list = [x1, x3, x4]
y_list = [y1, y3, y4]
color_indices = [0, 5, 2]  # Color indices for each curve


# Call the plotting function
plot_multiple_curves(x_list, y_list, color_indices, plot_title='V_Mitte')

df_wb1.columns
```
