## Same params of 1. tests series shells

```python
# params_2_cells_span_1800_height_300_width_500_a100
a = 100.0
b=493.478
c=188.083
gamma= 0.923173
n_x= 3
```

```python
# We can control gamma with our foldwork (only for gammas > 0.92)
gamma = 1.05 # presented in TRR 11.2022

wb_shell = WBTessellation4P(gamma = gamma, a=a, b=b, c=c,
                            n_x_plus=2,
                            n_phi_plus=6,
                         wireframe_width=5,
                         trim_half_cells_along_y=True,
                         trim_half_cells_along_x=True,
                         align_outer_nodes_along_x=True,)
wb_shell.interact()
```

```python
wb_shell = WBTessellation5PBeta(wb_cell=WBCell5ParamBeta(gamma = gamma, eta=b/a, zeta=c/a, delta_beta= 0),
                              sol_num=4, n_x=n_x, n_y=3)
wb_shell.interact()
```
