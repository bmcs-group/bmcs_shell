# Deriving $b$ so that $\gamma = \theta$ as shown

This makes two WB cells/slabs compatible when rotated 90



```python
%matplotlib widget
from bmcs_shell.api import WBCell4Param
import numpy as np
import sympy as sp
```

```python
wb_cell = WBCell4Param(gamma = np.pi/4)
wb_cell.interact()
```

```python
c = wb_cell.symb.c
gamma = wb_cell.symb.gamma
a = wb_cell.symb.a
b = wb_cell.symb.b

z_of_corner_vertices = wb_cell.symb.u_3_
b_where_gamma_equal_theta = sp.solve(sp.cos(gamma) - z_of_corner_vertices / b, b)
b_where_gamma_equal_theta
```
